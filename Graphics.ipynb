```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
motor0_dados1 = pd.read_csv('https://raw.githubusercontent.com/SilvaEder/PPGEE/main/CSV/motor0_data1.csv',encoding='utf-8')
motor1_dados1 = pd.read_csv('https://raw.githubusercontent.com/SilvaEder/PPGEE/main/CSV/motor1_data1.csv',encoding='utf-8')

motor0_median = pd.read_json('https://raw.githubusercontent.com/SilvaEder/PPGEE/main/JSON/motor0_median.json',encoding='utf-8')
motor1_median = pd.read_json('https://raw.githubusercontent.com/SilvaEder/PPGEE/main/JSON/motor1_median.json',encoding='utf-8')

motor0_func = pd.read_json('https://raw.githubusercontent.com/SilvaEder/PPGEE/main/JSON/motor0.json',encoding='utf-8')
motor1_func = pd.read_json('https://raw.githubusercontent.com/SilvaEder/PPGEE/main/JSON/motor1.json',encoding='utf-8')

motor0_resi = pd.read_json('https://raw.githubusercontent.com/SilvaEder/PPGEE/main/JSON/motor0_resi.json',encoding='utf-8')
motor1_resi = pd.read_json('https://raw.githubusercontent.com/SilvaEder/PPGEE/main/JSON/motor1_resi.json',encoding='utf-8')
```

```python
motor0_dados1.columns = ["sample","force","u"]
motor1_dados1.columns = ["sample","force","u"]
```

```python
plt.figure(figsize=(9, 4))
plt.plot(motor0_dados1['u'],motor0_dados1['force'],'b')
plt.title("Scatter Plot")
plt.ylabel("Force (N)")
plt.xlabel("Command Input (u)")
plt.grid(True)
plt.savefig('Figure/Pyplot01.png')
plt.show()
```

```python
motor0_median=np.asarray(motor0_median)
motor0_median[0]=sorted(motor0_median[0])
x=np.arange(0,10.5,.5)
y=motor0_median[0]

plt.figure(figsize=(9, 4))
plt.plot(x,y,'bo', label='Median')
plt.title("Scatter Plot")
plt.ylabel("Force (N)")
plt.xlabel("Command Input (u)")
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(9, 4))
x1=np.asarray(motor0_dados1['u'] .tolist())
y1=(motor0_func[0][1])+(motor0_func[1][1])*x1+(motor0_func[2][1])*x1**2+(motor0_func[3][1])*x1**3
plt.plot(x1,y1,'r', label='Model ')
plt.plot(x,y,'bo', label='Median')
plt.title("Modelo dos Dados Aquisitados")
plt.ylabel("Force (N)")
plt.xlabel("Command Input (u)")
plt.grid(True)
plt.legend()
plt.show()
```

```python
motor0_resi=np.asarray(motor0_resi)
x=np.arange(0,len(motor0_resi[0]),1)
y=motor0_resi[0]
plt.figure(figsize=(9, 4))
plt.plot(x,y)
plt.plot(x,y,'bo')
plt.axhline(y = 0, color = 'r', linestyle = '--')
plt.title("Distribuio dos Resduos")
plt.ylabel("Error")
plt.xlabel("Sample")
plt.show()
```
